### DESCRIPCION DE LA ACTIVIDAD


La tienda de materiales "Shaman Tools" desea implementar un sistema de inventario para gestionar la
venta de productos y materiales de construcción. Los administradores de la tienda proponen que el
sistema cuente con dos tipos de usuarios:
- Usuarios tipo "admin": tienen privilegios para realizar las siguientes acciones en el sistema:
o Registrar a otros usuarios (nombre, contraseña y rol).
o Agregar o actualizar productos en el inventario.
o Consultar el inventario (buscando por código o listando todos los productos
disponibles).
o Comprar productos para la tienda (si el producto ya existe, se aumenta la cantidad
disponible; si no, se agrega al inventario).
o Ver un resumen de las ventas realizadas (incluyendo productos vendidos, cantidad, y
total generado).
- Usuarios tipo "vendedor": solo pueden realizar las siguientes operaciones:
o Consultar productos (por código o listando todos los productos disponibles).
o Vender productos (indicando el código del producto y la cantidad vendida). El sistema
debe disminuir la cantidad en el inventario correspondiente.

1. Iniciar.
2. Mostrar la pantalla inicial del sistema de inventario.
3. Menú Principal:
Opción 1: Login (acceso al sistema).
Opción 2: Salir (cierra el sistema).
4. Login:

- Solicitar al usuario nombre y contraseña.
- Verificar las credenciales y rol del usuario:
-- Si las credenciales son correctas, determinar si es "admin" o "vendedor".
-- Si las credenciales son incorrectas, mostrar un mensaje de error y repetir el login.
5. Mostrar el Menú según el Rol:
- Menú Admin:

-- Opción 1: Registrar usuario.
-- Opción 2: Agregar/actualizar producto.
-- Opción 3: Consultar inventario (por código o lista completa).
-- Opción 4: Comprar productos.
-- Opción 5: Ver resumen de ventas.
-- Opción 6: Salir (volver al menú principal).

- Menú Vendedor:

Opción 1: Consultar productos.
Opción 2: Vender productos.
Opción 3: Salir (volver al menú principal).
6. Operaciones del Menú:

- Registrar Usuario (Admin):
Ingresar nombre, contraseña y rol.
Validar y agregar el usuario a la base de datos.

- Agregar/Actualizar Producto (Admin):
Ingresar el código, nombre, precio y cantidad del producto.
Si el producto existe, actualizar la cantidad; si no, agregarlo como nuevo.

- Consultar Inventario:
Buscar un producto por código o listar todos los productos disponibles.

- Comprar Productos (Admin):
Ingresar el código del producto y la cantidad que se va a comprar.
Si el producto ya existe, aumentar la cantidad; si no, agregarlo al inventario.

- Ver Resumen de Ventas (Admin):
Listar los productos vendidos, cantidad y total generado por ventas.

- Vender Productos (Vendedor):
Ingresar el código y cantidad del producto.
Verificar la disponibilidad en inventario y disminuir la cantidad vendida.
Registrar la venta en la base de datos.

7. Una vez completada la acción seleccionada, mostrar nuevamente el menú según el rol del usuario hasta que seleccione la opción de "Salir".
